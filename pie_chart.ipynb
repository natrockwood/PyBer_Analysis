```python
import matplotlib.pyplot as plt
```

```python
categories = ["pants", "shirts", "outerwear", "pajamas", "underwear", "shoes"]
```

```python
revenue = [25, 15,5,20,10,25]
```

```python
# pritning pie chart basic
plt.pie(revenue)
plt.show()
```

```python
# printing pie chart with labels from string
plt.pie(revenue, labels=categories)
plt.show()
```

```python
# printing pie chart with labels from string, autopct = % with one decimal point
plt.pie(revenue, labels=categories, autopct="%1.1f%%")
plt.show()
```

```python
# exploding a piece of the pie

#create list of values between 0 & 1 with the same size as ctegory list
explode = [0, .1, 0, 0, 0, 0]
plt.pie(revenue, labels=categories, autopct="%1.1f%%", explode = explode)
plt.show()
```
